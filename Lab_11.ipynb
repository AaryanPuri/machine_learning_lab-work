```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split 
from sklearn.metrics import confusion_matrix, classification_report


```

```python
df=pd.read_csv(r'C:\Users\acer\Desktop\SEM 6 PPTS\MLL\Company_Data.csv')
```

```python
lbl_enc=LabelEncoder()
df['ShelveLoc']=lbl_enc.fit_transform(df['ShelveLoc'])
df['Urban']=lbl_enc.fit_transform(df['Urban'])
df['US']=lbl_enc.fit_transform(df['US'])
```

```python
x=df.drop('US',axis=1)
```

```python
y=df['US']
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,shuffle=True,random_state=42)
```

```python
std=StandardScaler()
Xtrain=std.fit_transform(Xtrain)
Xtest=std.transform(Xtest)
```

```python
gnb=GaussianNB()
gnb.fit(Xtrain,ytrain)
```

```python
print("Training accuracy: ",gnb.score(Xtrain,ytrain))
print("Training accuracy: ",gnb.score(Xtest,ytest))
```

```python
y_pred=gnb.predict(Xtest)
```

```python
Xtest.shape[0]
```

```python
print("Number of mislabeled points out of a total %d points: %d" % (Xtest.shape[0], (ytest != y_pred).sum()))
```

```python
cm=confusion_matrix(ytest,y_pred)
print(cm)
```

```python
cr=classification_report(ytest,y_pred)
print(cr)
```
