```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\acer\Desktop\SEM 6 PPTS\MLL\tvmarketing.csv')
```

```python
df.plot(x='TV',y='Sales',kind='scatter')
```

#Linear Regression
y=c+m1x1+m2x2+...+mnxn
y(sales)=c+w1*TV

```python
x= df['TV']
y= df['Sales']
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(Xtrain.shape, ytrain.shape)
print(Xtest.shape,ytest.shape)
```

```python
lr=LinearRegression()
```

```python
Xtrain=np.expand_dims(Xtrain,1)
Xtest=np.expand_dims(Xtest,1)
```

```python
Xtrain = np.squeeze(Xtrain)
Xtest = np.squeeze(Xtest)
```

```python
print(Xtrain.shape)
print(Xtest.shape)
```

```python
(160,1)
```

```python
print("Reshaped training data shape:", Xtrain.shape)
print("Reshaped test data shape:", Xtest.shape)
```

```python
lr.fit(Xtrain,ytrain)
```

```python
print(lr.intercept_)
print(lr.coef_)
```

```python
# y(sales)=0.046+X(TV)+7.29
```

```python
lr.score(Xtrain,ytrain)
```

```python
y_pred=lr.predict(Xtest)
```

```python
c = [i for i in range(1,41,1)]   #generating index
fig = plt.figure()
plt.plot(c,ytest, color="blue", linewidth=2, linestyle="-")
plt.plot(c,y_pred, color="red", linewidth=2, linestyle="-")
fig.suptitle('Actual and Predicted', fontsize=20)  #Plot heading
plt.xlabel('Index', fontsize=18)  #X-label
plt.ylabel('Index', fontsize=16)  #Y-label
```

```python
from sklearn.metrics import mean_squared_log_error,r2_score

mse=mean_squared_log_error(ytest,y_pred)
print(mse)
```

```python
r2s=r2_score(ytest,y_pred)
print(r2s)
```
